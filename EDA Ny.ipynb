# Data pre processing

```python
## fix missing values
df.fillna(method='ffill', inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
# 1â€“6 = mild, 7â€“12 = moderate, 13â€“24 = severe
def categorize_k6(value):
    if value >= 0 and value <= 6:
        return 1 #
    elif value >= 7 and value <= 12:
        return 2 #moderate
    else: return 3
```

```python
df['k6_category'] = df['k6'].apply(categorize_k6)
```

```python
sns.countplot(df, x="k6_category")
```

## balance the data set

```python
df['k6_category'].value_counts()
```

# Model Building and Evaluations

```python
df_encoded.columns
```

```python
df_encoded['k6_category'].unique()
```

```python
X = df_encoded[['birthsex', 'insured', 'delaypayrent', 'everhivtest',
       'sexuallyactive', 'drinker', 'bmi', 'usborn', 'nsugardrinkperday',
       'exercise', 'toldhighbp', 'mood1_1',
       'mood1_2', 'mood1_3', 'mood1_4', 'mood1_5', 'agegroup_1.0',
       'agegroup_2.0', 'agegroup_3.0', 'agegroup_4.0', 'imputed_povgroup3_1',
       'imputed_povgroup3_2', 'imputed_povgroup3_3', 'generalhealth_1.0',
       'generalhealth_2.0', 'generalhealth_3.0', 'generalhealth_4.0',
       'generalhealth_5.0', 'smoker_1.0', 'smoker_2.0',
       'smoker_3.0', 'education_1.0', 'education_2.0', 'education_3.0',
       'education_4.0','employment_1.0', 'employment_2.0', 'employment_3.0', 'employment_4.0',
       'employment_5.0', 'employment_6.0', 'employment_7.0', 'employment_8.0']]
y = df_encoded['k6_category']
```

```python
from sklearn import linear_model
from sklearn.ensemble import RandomForestClassifier
import statsmodels.api as sm
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay, f1_score, classification_report
from sklearn.model_selection import RandomizedSearchCV, train_test_split
```

```python
def build_and_test_models(X_train, X_test, y_train, y_test):
    dic = {
    'Logistic Regression' : linear_model.LogisticRegression(max_iter=5000),
    'Random Forest': RandomForestClassifier(),
    'SGD Classifier': linear_model.SGDClassifier()
    }

    for name, model in dic.items():
        print('Training...', name )
        model.fit(X_train, y_train)
        prediction = model.predict(X_test)
        accuracy = accuracy_score(y_test, prediction)
        F1 = f1_score(y_test, prediction, average='micro')
        print('accuracy', accuracy, ' F1: ', F1)
        # Create the confusion matrix
        cm = confusion_matrix(y_test, prediction)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm)
        print(classification_report(y_test, prediction))
        disp.plot()
        plt.title(name)  # You can customize the title text here
        # Show the plot
        plt.show()
```

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
build_and_test_models(X_train, X_test, y_train, y_test)
```
